```python
import pandas as pd
#worldModel.plot_metrics(metricsDf, len(worldModel.agents), worldModel.n_tasks)
import seaborn as sns
import matplotlib as mpl

dfD = pd.read_csv('Resultado_Final_Qualify_Principal01_DYN.csv')
dfD['dynamic'] = True

dfS = pd.read_csv('Resultado_Final_Qualify_Principal01.csv')
dfS['dynamic'] = False

df = pd.concat([dfD, dfS])

df.F_distance = df.F_distance * 10 
df.F_time = df.F_time / 5 
#df.F_quality = df.F_quality * 10 

summary_table = df[['Algorithm', 'dynamic', 'F_time', 'F_distance', 'F_load', 'F_quality', 'F_Reward', 'S_Reward']].groupby(['Algorithm', 'dynamic']).agg(['mean', 'std'])
df


from bioinfokit.analys import stat

# Realizar o teste de Tukey-HSD
res = stat()
res.tukey_hsd(df=df, res_var='F_Reward', xfac_var=['Algorithm', 'dynamic'], anova_model='F_time~Algorithm*dynamic')

print(res.tukey_summary)


```

```python

#%%%
# Calculate the mean for each factor grouped by 'Algorithm' and 'dynamic'
mean_df = df.groupby(['Algorithm', 'dynamic']).mean().reset_index()

# Separate the dataframes for `dynamic = True` and `dynamic = False`
df_true = mean_df[mean_df['dynamic'] == True]
df_false = mean_df[mean_df['dynamic'] == False]

# Merge these two dataframes on 'Algorithm'
merged_df = pd.merge(df_false, df_true, on='Algorithm', suffixes=('_False', '_True'))

# Calculate the percentage difference for each factor
for column in ['F_time', 'F_distance', 'F_load', 'F_quality', 'F_Reward', 'S_Reward']:
    merged_df[column + '_True'] = ((merged_df[column + '_True'] - merged_df[column + '_False']) / merged_df[column + '_False']) * 100

# Drop the '_False' columns as they are no longer needed
#merged_df.drop(columns=[col for col in merged_df.columns if '_False' in col], inplace=True)

# Rename the '_True' columns to remove the suffix
merged_df.rename(columns={col: col.replace('_True', '') for col in merged_df.columns}, inplace=True)

merged_df
```
